```python

from skimage.color import rgb2gray
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

##  灰度化

### 方法1

```python
# 灰度化
img = cv2.imread("BB.jpg")

img_gray = rgb2gray(img)

plt.imshow(img_gray, cmap='gray')
```

### 方法2

```python

img_gray2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#plt.imshow(img_gray2, cmap='gray')

#CV2 
cv2.imshow("gray_image",img_gray2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 方法3

```python
img = cv2.imread("BB.jpg")
h,w = img.shape[:2]
img_gray3 = np.zeros([h,w],img.dtype)
```

```python
for i in range(h):
    for j in range(w):
         m = img[i,j]          
         img_gray3[i,j] = int(m[0]*0.114 + m[1]*0.587 + m[2]*0.299)
       
cv2.imshow("gray",img_gray3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 二值化

```python
img = cv2.imread("BB.jpg")

img_gray4 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#先转为灰度图，再二值化


img_binary = np.where(img_gray4 >= np.average(img_gray4), 255, 0) 


plt.imshow(img_binary, cmap='gray')
```
